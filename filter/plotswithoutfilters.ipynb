```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(filename):
    with open(filename) as json_file:
        data = json.load(json_file)
        return np.array(data)
```

```python
data = load_data("../data/в покое.json")
```

```python
def data_taking(data):
    vectors_accelerometer_x = []
    vectors_accelerometer_y = []
    vectors_accelerometer_z = []
    ox = []

    counter = 0
    for gesture in data:
        counter += 1
        if counter != 1: continue
        # for time_series in gesture:
        for i in range(1, len(gesture)):
            vectors_accelerometer_x.append(gesture[i][1])
            vectors_accelerometer_y.append(gesture[i][2])
            vectors_accelerometer_z.append(gesture[i][3])
            ox.append(gesture[i][0])
    return ox, vectors_accelerometer_x, vectors_accelerometer_y, vectors_accelerometer_z

ox, vectors_accelerometer_x, vectors_accelerometer_y, vectors_accelerometer_z = data_taking(data)


#print(len(ox))
#print(ox)
#print(vectors_accelerometer_x)
```

### В начале рассмотрим сырые данные без фильтрации, чтобы интуитивно понимать где возмоно будут проблемы при дальнейших преобразованиях

```python
def draw_ax(data_ax, name_ax, color):
    plt.figure(figsize=(16, 5))
    plt.plot(ox, data_ax, color + '-')
    plt.style.use('fivethirtyeight')
    plt.title('Данные с акселерометра по оси ' + name_ax + ' без фильтрации в покое')
    plt.ylabel('значения в м/с^2')
    plt.xlabel('время в миллисекундах')
    plt.show()
```

```python
draw_ax(vectors_accelerometer_x, 'X', 'r')
```

```python
draw_ax(vectors_accelerometer_y, 'Y', 'b')
```

```python
draw_ax(vectors_accelerometer_z, 'Z', 'g')
```

```python
def draw_ax_circle(data_ax, name_ax, color, num, ox_c):
    plt.figure(figsize=(16, 5))
    plt.plot(ox_c, data_ax, color + '-')
    plt.style.use('fivethirtyeight')
    plt.title('Данные с акселерометра по оси ' + name_ax + ' без фильтрации круг ' + str(num))
    plt.ylabel('значения в м/с^2')
    plt.xlabel('время в миллисекундах')
    plt.show()
```

```python
draw_ax_circle(vectors_accelerometer_x_c, 'X', 'r', 1, ox_c)
```

```python
draw_ax_circle(vectors_accelerometer_y_c, 'Y', 'b', 1)
```

```python
draw_ax_circle(vectors_accelerometer_z_c, 'Z', 'g', 1)
```

```python
def draw_all_circles(num):
    for num_c in range(2, num + 1):
        temp_data = load_data('../data/круг ' + str(num_c) + '.json')
        ox_t, vectors_accelerometer_x_t, vectors_accelerometer_y_t, vectors_accelerometer_z_t = data_taking(temp_data)
        str_oxes = ['X', 'Y', 'Z']
        colors = ['r', 'b', 'g']
        vecs_oxes = [vectors_accelerometer_x_t, vectors_accelerometer_y_t, vectors_accelerometer_z_t]
        for i in range(3):
            draw_ax_circle(vecs_oxes[i], str_oxes[i], colors[i], num_c, ox_t)
```

```python
draw_all_circles(5)
```

```python
def draw_ax_square(data_ax, name_ax, color, num, ox_c):
    plt.figure(figsize=(16, 5))
    plt.plot(ox_c, data_ax, color + '-')
    plt.style.use('fivethirtyeight')
    plt.title('Данные с акселерометра по оси ' + name_ax + ' без фильтрации угол' + str(num))
    plt.ylabel('значения в м/с^2')
    plt.xlabel('время в миллисекундах')
    plt.show()

square = load_data("../data/квадрат.json")
ox_s, vectors_accelerometer_x_s, vectors_accelerometer_y_s, vectors_accelerometer_z_s = data_taking(square)
draw_ax_square(vectors_accelerometer_x_s, 'X', 'r', 1, ox_s)
draw_ax_square(vectors_accelerometer_y_s, 'Y', 'b', 1, ox_s)
draw_ax_square(vectors_accelerometer_z_s, 'Z', 'g', 1, ox_s)
```
